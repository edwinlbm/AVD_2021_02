```python
itens = ['baby', 'tea', 'biscuits', 'coffee', 'frozen', 'spices', 'insecticides','pet', 'soft', 'haircare', 'dental', 'cheese', 'chickens', 'margarine', 'salads', 'juice-sat-cord-ms', 'beef', 'pork',  'fruit.1', 'vegetables'] 
```

```python
dfAmostraSupermarket = dfSupermarket[itens]
dfAmostraSupermarket
```

```python
listaCompras = []
for linha in dfAmostraSupermarket.values:
  lista = []
  for l in range(len(linha)):
    if linha[l] == 1:
      lista.append(dfAmostraSupermarket.columns[l])
  listaCompras.append(lista)

dfListaCompras  = pd.DataFrame(listaCompras)
dfListaCompras
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(listaCompras).transform(listaCompras)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

##Apriori

```python
from mlxtend.frequent_patterns import apriori

itemsets_apriori = apriori(df, min_support=0.3, use_colnames=True)
itemsets_apriori
```

##Fpgrowth

```python
from mlxtend.frequent_patterns import fpgrowth as mlxtendfpgrowth

itemsets_fpgrowth = mlxtendfpgrowth(df, min_support=0.3, use_colnames=True)
itemsets_fpgrowth
```

##Regras de Associação

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(itemsets_apriori, metric="confidence", min_threshold=0.7)
```

```python
association_rules(itemsets_fpgrowth, metric="confidence", min_threshold=0.7)
```
